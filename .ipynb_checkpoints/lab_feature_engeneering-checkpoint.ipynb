````python
import pandas as pd
import numpy as np


import warnings
warnings.filterwarnings('ignore')


import matplotlib.pyplot as plt
%matplotlib inline


import seaborn as sns
````

````python
data = pd.read_csv("learningSet.txt")
data
````

````python
# Lets create a drop_list we will use later as well
drop_list = list(data[['OSOURCE', 'ZIP']])
drop_list
````

````python
# Lets deal with sparcity part 
# Lets check null values in percentage 

nulls_percent_df= data.isna().sum()/len(data)
nulls_percent_df

# put it in a dataframe 
nulls_percent_df= pd.DataFrame(data.isna().sum()/len(data))
nulls_percent_df

# Take out of the index 
nulls_percent_df= pd.DataFrame(data.isna().sum()/len(data)).reset_index()
nulls_percent_df

# Lets change columns name
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df

# Lets sort 
nulls_percent_df.sort_values(by = ['nulls_percentage'], ascending = False)
````

````python
# First create the variable with the threshold 
threshold =0.25 

# define a condition 
condition = nulls_percent_df['nulls_percentage']>threshold
columns_above_threshold = nulls_percent_df[condition]
columns_above_threshold

# Create a list with column names
drop_columns_list = list(columns_above_threshold['column_name'])
drop_columns_list
````

````python
# Removing the null values with a threshold above 25% using the list created above that shows all the columns within this th
data_drop1 = data.drop(columns=drop_columns_list)
data_drop1
````

````python
# Check and fill the null values with F in the GENDER column
print(data['GENDER'].value_counts())
data['GENDER'] = data['GENDER'].fillna('F')
````

````python
# Lets settle the values of GENDER to only M F or other 
def frequent_values(df, column, n=2, replace_value='other'):
    value_counts = df[column].value_counts()
    top_n_values = value_counts.index[:n]
    df[column] = df[column].apply(lambda x: x if x in top_n_values else replace_value)

    return df
````

````python
frequent_values(data, 'GENDER', n=2, replace_value='other')
````

````python
numerical_data = data.select_dtypes(include = np.number)
numerical_data
````

````python
# Checking for null values in numerical_data DF
nulls_percent_numerical= pd.DataFrame(numerical_data.isna().sum()/len(numerical_data)).reset_index()
nulls_percent_numerical.columns = ['column_name', 'nulls_percentage']

# Sorting the values to see the highest first
nulls_percent_numerical.sort_values(by = ['nulls_percentage'], ascending = False)
````

````python
# Variables to treat GEOCODE2, WEALTH1, ADI, DMA,and MSA.
data['GEOCODE2'].value_counts()
# In here we can see that there is a a blanc variables, but is not a NAN otherwise it ould appear NAN, so I will just turn the blanc value into the most common value A
data['GEOCODE2'].replace(' ', 'A', inplace = True)
data['GEOCODE2'].value_counts()

#lets plot this variable first and check the null values 
data['GEOCODE2'].isna().sum()

sns.histplot(data['GEOCODE2'])
plt.show()

# Since there is only 132, since this is a discrete varibale I will just fill it with the mode
data['GEOCODE2']= data['GEOCODE2'].fillna('A')
data['GEOCODE2'].isna().sum()
````

````python
# For the Wealth1 variable 
#First check frequency and if it is th write type and it is on float
print(data['WEALTH1'].value_counts())
print(data['WEALTH1'].dtypes)

# Check null values 
print(data['WEALTH1'].isna().sum())

# There is almost 50% of null values in this column so I will just drop it 

data = data.drop(columns = 'WEALTH1', axis=1)
data
````

````python
# ADI variable
#First check frequency and if it is th write type and it is on float
print(data['ADI'].value_counts())
print(data['ADI'].dtypes)

# Check null values 
print(data['ADI'].isna().sum())

sns.distplot(data['ADI'])
plt.show()

# Since this is a continuous variable and we only have 132 null values I will fill them with the mean
data['ADI'] = data['ADI'].fillna(np.mean(data['ADI']))

# Check if the distribution changed
sns.distplot(data['ADI'])
plt.show()

# Check if all the null values were actually filled
print(data['ADI'].isna().sum())
````

````python
# DMA variable
#First check frequency and if it is th write type and it is on float
print(data['DMA'].value_counts())
print(data['DMA'].dtypes)

# Check null values 
print(data['DMA'].isna().sum())

sns.distplot(data['DMA'])
plt.show()

# Since this is a continuous variable and we only have 132 null values I will fill them with the mean
data['DMA'] = data['DMA'].fillna(np.mean(data['DMA']))

# Check if the distribution changed
sns.distplot(data['DMA'])
plt.show()
````

````python
# MSA variable
#First check frequency and if it is th write type and it is on float
print(data['MSA'].value_counts())
print(data['MSA'].dtypes)

# Check null values 
print(data['MSA'].isna().sum())

sns.distplot(data['MSA'])
plt.show()

# Since this is a continuous variable and we only have 132 null values I will fill them with the mean
data['MSA'] = data['MSA'].fillna(np.mean(data['MSA']))

# Check if the distribution changed
sns.distplot(data['MSA'])
plt.show()
````
