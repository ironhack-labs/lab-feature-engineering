```python
# Libraies
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exercise: Clean dataset
# 1. Use appropriate methods to clean the columns GEOCODE2, WEALTH1, ADI, DMA,and MSA.
```

```python
df = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\15_Intro_ML\learningSet_3.txt')
df
# 95.412 x 385
```

```python
# Column 'GEOCODE2'
df['GEOCODE2'].value_counts()

# Filling NaNs (' ') with the mode ('A') 
df['GEOCODE2'].replace(' ', 'A', inplace = True)
df['GEOCODE2'].value_counts()
```

```python
# 'WEALTH1' does not exist in the df
```

```python
# Column 'ADI'
df['ADI'].unique
sns.distplot(df['ADI'])
plt.show()
df['ADI'].isna().sum()

# Fillining NaNs with the mean
df['ADI'] = df['ADI'].fillna(np.mean(df['ADI']))
sns.distplot(df['ADI'])
plt.show()
```

```python
# Column 'DMA'
df['DMA'].unique
sns.distplot(df['DMA'])
plt.show()
df['DMA'].isna().sum()

# Fillining NaNs with the mean
df['DMA'] = df['DMA'].fillna(np.mean(df['DMA']))
sns.distplot(df['DMA'])
# plt.show()
```

```python
# Column 'MSA'
df['MSA'].unique
sns.distplot(df['MSA'])
plt.show()
df['MSA'].isna().sum()

# Fillining NaNs with the mean
df['MSA'] = df['MSA'].fillna(np.mean(df['MSA']))
sns.distplot(df['MSA'])
plt.show()
```

```python
# 2. Dealing with the rest of numeric columns
sum_na = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
sum_na.columns = ['column_name', 'nans_percentage']
sum_na.sort_values(by = ['nans_percentage'], ascending = False).head(10)
```
