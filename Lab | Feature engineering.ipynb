```python
import pandas as pd
import numpy as np
```

```python
categorical = pd.read_csv('categorical_df.csv',index_col = 0)
numerical = pd.read_csv('numerical_df.csv',index_col = 0)
Y = pd.read_csv('Y.csv')
```

```python
numerical
```

```python
df = pd.DataFrame(numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
numerical['WEALTH1'].value_counts()
```

```python
numerical['WEALTH1'] = numerical['WEALTH1'].fillna('9.0')
```

```python
numerical['ADI'].value_counts()
```

```python
numerical['ADI'] = numerical['ADI'].fillna(round(np.mean(numerical['ADI'])))
```

```python
numerical['DMA'].value_counts()
```

```python
numerical['DMA'] = numerical['DMA'].fillna(round(np.mean(numerical['DMA'])))
```

```python
numerical['MSA'].value_counts()
```

```python
numerical['MSA'] = numerical['MSA'].fillna(round(np.mean(numerical['MSA'])))
```

```python
df = pd.DataFrame(numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
numerical.drop(['NEXTDATE'],axis=1,inplace=True)
```
