```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
numerical = pd.read_csv('numerical.csv')
categorical = pd.read_csv('categorical.csv')
target = pd.read_csv('Y.csv')
```

```python
df = pd.DataFrame(numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
sns.histplot(numerical['WEALTH1'])
```

```python
numerical['WEALTH1'] = numerical['WEALTH1'].fillna(round(numerical['WEALTH1'].mode().values[0]))
```

```python
sns.histplot(numerical['ADI'])
```

```python
numerical['ADI'] = numerical['ADI'].fillna(round(numerical['ADI'].mean()))
```

```python
sns.histplot(numerical['DMA'])
```

```python
numerical['DMA'] = numerical['DMA'].fillna(round(numerical['DMA'].mean()))
```

```python
sns.histplot(numerical['MSA'])
```

```python
numerical['MSA'] = numerical['MSA'].fillna(round(numerical['MSA'].mode().values[0]))
```

```python
numerical.drop(['NEXTDATE'],axis=1,inplace=True)
```

```python
df_categorical = pd.DataFrame(categorical.isna().sum()).reset_index()
df_categorical.columns = ['column_name', 'nulls']
df_categorical[df_categorical['nulls']>0]
```

```python
sns.histplot(categorical['GEOCODE2'])
```

```python
categorical.drop(['OSOURCE'],axis=1,inplace=True)
categorical.drop(['GENDER'],axis=1,inplace=True)
categorical.drop(['SOLIH'],axis=1,inplace=True)
categorical.drop(['VETERANS'],axis=1,inplace=True)
```

```python
numerical.to_csv('numerical.csv')
```

```python
categorical.to_csv('categorical.csv')
```

```python
target.to_csv('Y.csv')
```
