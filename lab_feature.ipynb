```python
import pandas as pd
```

```python
data_numerical['WEALTH1'].value_counts()
```

```python
#Filling with the mode - WEALTH1
```

```python
data_numerical['WEALTH1'] = data_numerical['WEALTH1'].fillna(9.0)
```

```python
data_numerical['ADI'].value_counts()
```

```python
data_numerical['DMA'].value_counts()
```

```python
data_numerical['MSA'].value_counts()
```

```python
#The other 3 columns filling with the mean
```

```python
data_numerical['ADI'] = data_numerical['ADI'].fillna(round(data_numerical['ADI'].mean()))
```

```python
data_numerical['DMA'] = data_numerical['DMA'].fillna(round(data_numerical['DMA'].mean()))
```

```python
data_numerical['MSA'] = data_numerical['MSA'].fillna(round(data_numerical['MSA'].mean()))
```

```python
df = pd.DataFrame(data_numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
data_numerical = data_numerical.drop(['NEXTDATE'], axis=1)
```

```python
df = pd.DataFrame(data_categorical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

```python
data_categorical = data_categorical.drop(['OSOURCE', 'GENDER', 'SOLIH', 'VETERANS', 'GEOCODE2'], axis=1)
```

```python
data_numerical.to_csv('numerical_clean_feature_engineering',index = False)
data_categorical.to_csv('categorical_clean_feature_engineering',index = False)
```
