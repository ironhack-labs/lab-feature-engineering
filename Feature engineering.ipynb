```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
categorical = pd.read_csv('categorical.csv')
numerical2 = pd.read_csv('numerical2.csv')
targets = pd.read_csv('targets.csv')
```

```python
df = pd.DataFrame(numerical2.isna().sum()/len(numerical2)).reset_index()
df.columns = ['column_name','nulls']
df[df['nulls']>0].sort_values(by='nulls',ascending=False)
```

```python
numerical2['WEALTH1'].value_counts()
```

```python
numerical2['WEALTH1'] = numerical2['WEALTH1'].fillna(round(numerical2['WEALTH1'].mode().values[0]))
```

```python
sns.histplot(numerical2['WEALTH1'])
```

```python
numerical2['NEXTDATE'].value_counts()
```

```python
numerical2['NEXTDATE'] = numerical2['NEXTDATE'].fillna(round(numerical2['NEXTDATE'].mean()))
```

```python
sns.histplot(numerical2['NEXTDATE'])
```

```python
numerical2['MSA'].value_counts()
```

```python
numerical2['MSA'] = numerical2['MSA'].fillna(numerical2['MSA'].mean())
```

```python
sns.histplot(numerical2['MSA'])
```

```python
numerical2['ADI'].value_counts() 
```

```python
numerical2['ADI'] = numerical2['ADI'].fillna(round(numerical2['ADI'].mean()))
```

```python
sns.histplot(numerical2['ADI'])
```

```python
numerical2['DMA'].value_counts() 
```

```python
numerical2['DMA'] = numerical2['DMA'].fillna(round(numerical2['DMA'].mean()))
```

```python
sns.histplot(numerical2['DMA'])
```
