```python
numerical['WEALTH1'] = numerical['WEALTH1'].fillna('9.0')
```

```python
numerical['ADI'].value_counts()
```

```python
sns.histplot(numerical['ADI'])
```

```python
numerical['ADI'] = numerical['ADI'].fillna(round(numerical['ADI'].mean()))
```

```python
numerical['DMA'].value_counts()
```

```python
numerical['DMA'] = numerical['DMA'].fillna(round(numerical['DMA'].mean()))
```

```python
numerical['MSA'].value_counts()
```

```python
numerical['MSA'] = numerical['MSA'].fillna(round(numerical['MSA'].mean()))
```

```python
data = pd.DataFrame(numerical.isna().sum()).reset_index() 
data.columns = ['columns_name', 'nulls']
data[data['nulls'] > 0]
```

```python
numerical = numerical.drop(['NEXTDATE'], axis=1)
```

```python
data = pd.DataFrame(numerical.isna().sum()).reset_index() 
data.columns = ['columns_name', 'nulls']
data[data['nulls'] > 0]
```

```python
# no null values in numerical
```

```python
data = pd.DataFrame(categorical.isna().sum()).reset_index() 
data.columns = ['columns_name', 'nulls']
data[data['nulls'] > 0]
```
